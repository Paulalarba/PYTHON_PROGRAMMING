```python
#ChatGPT Code (paste verbatim)
import random

def roll() -> int:
    """
    Simulates rolling a fair six-sided die.
    Returns an integer from 1 to 6 inclusive.
    """
    return random.randint(1, 6)

def main():
    simulations = 100_000
    target_sum = 7
    count = 0

    # Optional: set a random seed for reproducibility
    # random.seed(42)

    for _ in range(simulations):
        die1 = roll()
        die2 = roll()
        if die1 + die2 == target_sum:
            count += 1

    empirical_prob = count / simulations
    exact_prob = 6 / 36  # 6 ways to get sum 7 out of 36 total outcomes

    print(f"After {simulations} simulations:")
    print(f"Empirical probability of sum {target_sum}: {empirical_prob:.4f}")
    print(f"Exact mathematical probability: {exact_prob:.4f}")

    print("\nNote:")
    print(" - Monte Carlo simulations rely on pseudo-random numbers.")
    print(" - Setting a seed (random.seed()) ensures reproducibility of results.")
    print(" - More simulations increase accuracy but take more computation time.")

if __name__ == "__main__":
    main()
```

```python
import random

def roll() -> int:
    """
    Simulates rolling a fair six-sided die.
    Returns an integer from 1 to 6 inclusive.
    """
    return random.randint(1, 6)

def monte_carlo_dice(target_sum: int, simulations: int) -> float:
    """
    Runs Monte Carlo simulations to estimate the probability
    that the sum of two dice equals target_sum.

    Returns the empirical probability.
    """
    if simulations <= 0:
        raise ValueError("Number of simulations must be a positive integer.")
    if simulations > 10_000_000:
        print("⚠️ Warning: very large number of simulations may slow down execution.")

    count = 0
    for _ in range(simulations):
        if roll() + roll() == target_sum:
            count += 1
    return count / simulations

def main():
    target_sum = 7

    try:
        simulations = int(input("Enter number of simulations (positive integer): "))
    except ValueError:
        print("Invalid input. Using default 100,000 simulations.")
        simulations = 100_000

    # Optional: set random seed for reproducibility
    # random.seed(42)

    empirical_prob = monte_carlo_dice(target_sum, simulations)
    exact_prob = 6 / 36  # 6 ways to get sum 7 out of 36 total outcomes

    print("\n Dice Simulation Results")
    print(f"Target sum: {target_sum}")
    print(f"Number of simulations: {simulations:,}")
    print(f"Empirical probability: {empirical_prob:.6f}")
    print(f"Exact probability:     {exact_prob:.6f}")
    print(f"Difference:            {abs(empirical_prob - exact_prob):.6f}")

    print("\nNote:")
    print(" - Monte Carlo simulations rely on pseudo-random numbers.")
    print(" - Setting a seed (random.seed()) ensures reproducibility.")
    print(" - More simulations generally increase accuracy but take longer.")

if __name__ == "__main__":
    main()
```
