```python
#ChatGPT Code (paste verbatim)
def normalize_and_find(text: str, needle: str) -> int:
    """
    Trims whitespace, converts to lowercase, and returns the index of
    the first occurrence of 'needle' in 'text'.
    Uses a safe search — returns -1 if not found.
    """
    # Step 1: Trim whitespace on both ends
    normalized_text = text.strip()

    # Step 2: Convert both text and needle to lowercase
    normalized_text = normalized_text.lower()
    needle = needle.lower()

    # Step 3: Safe search using find()
    return normalized_text.find(needle)


def main():
    # Input from user
    text = input("Enter the main text: ")
    needle = input("Enter the word to search for: ")

    # Perform search
    index = normalize_and_find(text, needle)

    # Display result
    if index != -1:
        print(f"'{needle}' found at index {index}.")
    else:
        print(f"'{needle}' not found in the text.")


# Standard Python entry point
if __name__ == "__main__":
    main()
```

```python
# Improved Code
def normalize_and_find(text: str, needle: str) -> int:
    """
    Trims whitespace, converts to lowercase, and returns the index of
    the first occurrence of 'needle' in 'text'.
    Uses a safe search — returns -1 if not found.
    """
    # Input validation
    if not isinstance(text, str) or not isinstance(needle, str):
        raise TypeError("Both 'text' and 'needle' must be strings.")

    # Step 1–2: Normalize
    normalized_text = text.strip().lower()
    needle = needle.strip().lower()

    # Step 3: Safe search using find()
    return normalized_text.find(needle)


def main():
    text = input("Enter the main text: ").strip()
    needle = input("Enter the word to search for: ").strip()

    if not needle:
        print("Search term cannot be empty.")
        return

    index = normalize_and_find(text, needle)

    if index != -1:
        print(f"'{needle}' found at index {index}.")
    else:
        print(f"'{needle}' not found in the text.")


if __name__ == "__main__":
    main()
```
