```python
from pathlib import Path
import shutil

base = Path.cwd()
my_folder = base / "my_folder"
my_folder.mkdir(exist_ok="True")


(file1 := my_folder / "file1.txt").touch(exist_ok="True")
(file2 := my_folder / "file2.txt").touch(exist_ok="True")
(image1 := my_folder / "image1.png").touch(exist_ok="True")

my_folder = my_folder / "images"
my_folder.mkdir(exist_ok="True")

shutil.move(str(image1), str(my_folder / image1.name))

file1.unlink(missing_ok=True)

shutil.rmtree(my_folder)
```

```python
from pathlib import Path

file_name = "starships.txt"
content = "\t Discovery \n\t Enterprise \n\t Defiant \n\t Voyager"

with open(file_name, 'w') as write:
    write.write(content)

with open('starships.txt', 'r') as read:
    text = read.read()
    print(text)

file = open('starships.txt', 'r')
for line in file:
    if line.strip().startswith('D'):
        print("\n", line, end='')
file.close()
    
```

```python
from pathlib import Path
import csv 

numbers = []

folder = Path.cwd()
file_path = folder / "numbers.csv"

numbers = []

with open(file_path, 'r') as read:
    reader = csv.reader(read)
    for row in reader:
        numbers.append([int(x) for x in row])

print("numbers= [")
for row in numbers:
    print(f"    {row}, ", end="")
print("\n]")
```

```python
from pathlib import Path
import csv

folder = Path.cwd()
file_path = folder / "Favorite_colors.csv"

favorite_colors = [
    {"name": "Joe", "favorite_color": "blue"},
    {"name": "Anne", "favorite_color": "green"},
    {"name": "Bailey", "favorite_color": "red"},
]

with open(file_path, 'w', newline="") as write:
    writer = csv.writer(write)
    writer.writerow(["name", "favorite_color"])

    for person in favorite_colors:
        writer.writerow([person["name"], person["favorite_color"]])
```

```python
from pathlib import Path
import csv

file_path = Path.cwd() / "Favorite_colors.csv"
favorite_colors = []

with open(file_path, 'r') as readfile:
    reader = csv.DictReader(readfile)

    for row in reader:

        favorite_colors.append({
            "name": row["name"],
            "favorite_color": row["favorite_color"]
        })

print("[")
for item in favorite_colors:
    print(f'{{"name": "{item["name"]}", "favorite_color": "{item["favorite_color"]}"}}, ')
print("]")
```
